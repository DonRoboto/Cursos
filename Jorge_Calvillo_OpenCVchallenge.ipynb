#Procesamiento de imágenes con OpenCV
En este ejercicio utilizarás los métodos de OpenCV para tratamiento de imágenes.
Leerás una imagen, reducirás su tamaño, convertirás a escala de grises, rotarás y más.

```
# EJERCICIO
# Importa las librerías que necesites para tratamiento de imágenes
import matplotlib.pyplot as plt
import numpy as np
from keras.preprocessing import image as image_utils
import cv2
import imutils
```

# Recuerda subir la imagen a colaboratory

```
# EJERCICIO
# Lee la imagen mediante el método de OpenCV y grafícala
imagen = cv2.imread('img_opencv.jpg')
imagen = cv2.cvtColor(imagen, cv2.COLOR_RGB2BGR)
plt.imshow(imagen)
plt.axis('on')
plt.show()
```

```
# EJERCICIO
# Escala la imagen original a un tamaño de 600 (ancho) x 300 (alto) pixeles y guardala en la variable imgSize
# Grafíca la imagen
imgSize = cv2.resize(imagen,(600,300), interpolation=cv2.INTER_CUBIC)
plt.imshow(imgSize)
plt.axis('on')
plt.show()
```

```
# EJERCICIO
# Rota la imagen original a 45° respetando los bordes
# Guarda la imagen resultante en imgRotate y grafícala
alto, ancho, canales = imagen.shape
M = cv2.getRotationMatrix2D((ancho/2, alto/2), 45, 1)
imgRotate = cv2.warpAffine(imagen, M, (ancho, alto))
plt.imshow(image_utils.array_to_img(imgRotate), cmap='Accent')
plt.show()
```

```
# EJERCICIO
# Cambia a escala de grises la imagen rotada
# Guarda la imagen resultante en imgGray y grafícala
imgGray = cv2.cvtColor(imgRotate, cv2.COLOR_BGR2GRAY)
plt.imshow(imgGray, cmap='gray')
plt.show()
```

```
# EJERCICIO
# Aplica ruido gaussiano a la imagen rotada
# Guarda la imagen resultante en imgNoise y grafíca la imagen
gaussian = np.zeros(imgRotate.shape)
cv2.randn(gaussian, 0.0, 40)
imgNoise = imgRotate + gaussian
plt.imshow(image_utils.array_to_img(imgNoise), cmap='Accent')
plt.show()
```

```
# EJERCICIO
# Aplica un filtro gaussiano a la imagen que previamente se generó con ruido
# Guarda la imagen resultante en imgClean y grafícala
imgClean = cv2.GaussianBlur(imgNoise, (5,7), 0)
plt.imshow(image_utils.array_to_img(imgClean), cmap='Accent')
plt.show()
```

```
# EJERCICIO
# Obten los contornos de la imagen mediante el algoritmo Canny
# Guarda la imagen resultante en imgCanny
# Grafíca la imagen
imgCanny = cv2.Canny(imagen.astype('uint8'), 150,200)
plt.imshow(imgCanny, cmap='gray')
plt.show()
```
